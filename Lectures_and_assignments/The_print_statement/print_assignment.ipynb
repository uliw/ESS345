```python
""" 
Program Name: bin_2_dec

command line parameters: none

Description: This program will take a variable containing a 
binary number (i.e., a sequence containing only zeros and ones)
and convert it into its decimal notation. The result will be 
printed on the screen.

Version: 0.1

Author: Uli Wortmann uli.wortmann@utoronto.ca

Copyright: 2019 Uli Wortmann

License: GPL
"""

# Parameters (i.e., variables you change for each run)
b = "01001" # string containig the binary number

# Scalar variables (i.e., single value only)
d = 0 # integer which holds the result of the conversion
n = 0 # counter which increase for each element of b_r
p = 0 # value which contains the 2^n

# Complex data types (strings, lists, vectors)
b_r = b[::-1] # b in reverse order
result = f""  # fstring for the results display

# ---------------------- code ----------------------------*

for i in b_r:     # loop over each element in b_r
    p = 2**n      # calculate 2**n
    if i == "1":
        d = d + p # add p to d if i != zero
    else:
        pass      # do nothing
    n = n + 1     # increase counter n

result = f"The binary number {b} equals {d} in decimal notation"
print(result)
```
