### Accessing list elements



#### Single list elements



Consider the following list:



```python
my_list=[1.12, 3.2, 1.45, 12.01, 1.12, 3.2, 1.45, 12.01]
```

In real life, lists are often rather long, and we only want to see
certain elements of it. Say you only want to see the second element of
this list, so we can write



```python
print(my_list[2])
```

The number in the square bracket is called the index, and refers to
position of the list element inside the list. Why do we see the 3<sup>rd</sup>
element of this list, rather then the 2<sup>nd</sup>?



#### Accessing a range of elements (slicing)



If we want to see several elements of a list (a so called slice), we
can specify a range



```python
print(my_list[2:4])
```

make sure you actually test the result of this operation by retrieving
each value alone (i.e., `print(my_list[2])` etc). What do you think is
happening here?



#### Accessing the last element(s) in a list



Sometimes, we have a-priory knowledge of how many list elements there
are. So python provides a short hand to access the last element:



```python
print(my_list[-1])    # the last element in the list
print(my_list[-2])    # the second last
print(my_list[-3:-1]) # the third and second last!
print(my_list[-3:])   # the last 3
```

#### Other list slices



In the previous example, we have seen that the colon separates the
beginning and end values of a list range, and that if we omit the
number, it will default to max index. The same is true for the
starting index



```python
print(my_list[0:3]) # will print the first 3 values
print(my_list[:3])  # this is the same
print(my_list[:])   # will print the whole list
```

last but not least, we can add a third argument to the range, to tell python to only access every n-th element.



```python
print(my_list[0:-1:2]) # every second element until the second last
print(my_list[0::2])   # every second element including the last
print(my_list[::2])    # the same, but shorter
```

and we can extend this to display the list in reverse. I.e., we
specify the higher index value as a start value, and the lower index
value as a stop value, and then we use a negative step size (this is
important!)



```python
print(my_list[-1:0:-1]) # show the list starting with the highest index
print(my_list[-1::-1])  # the same
```

#### Slicing: Take me home



-   In python, the first list element has the index of 0 (i.e., we
    start counting from zero). This is very different from matlab
    where the first element has an index value of 1.
    -   Specifying a range of elements using the colon, from say 1 to 5
        (`[1:5]`) means that we include the value at `[1]` but stop the
        range at `[4]`. In other words, this would state: Include all
        elements with an index equal or greater than one, but smaller
        than 5.  Since python lists are counted from zero, the above
        statement says: Include all values from index position 1 (which
        is the second list element) to index position 4 (which is the
        5<sup>th</sup> list element). Confusing? Yes, but you will get used to it!
    -   We can use the colon without a leading and trailing number,
        which simply implies that the range starts/stops with the
        first/last list element.
    -   We can add a step size "n" to the range expression so that we
        only see every n-th element
    -   We can use a negative step size to that we move through the list
        from the higher to the lower index.
