## Other Data Types



Python knows a variety of data types, many of which we may never
use. However, the most important ones should at least be mentioned:



### Vectors



Unlike lists, which contain almost anything, vectors can only contain
numbers and support mathematical methods (i.e., multiplying two
vectors will give you the cross product). We will use vectors in a
later module.  



### Tuples



 
Python tuples are a special version of a list, which won't allow you
to modify the value of a list element. We define a tuple by declaring
the list with regular brackets rather than square brackets.



```python
my_list = [1, 2, 4]       # a regular list
my_tuple_list = (1, 2, 3) # a tuple list
```

We access elements of a tuple via index or range operations similar to
regular list



```python
print(my_tuple_list[1]) # note the square brackets for the index expression!
```

however, you cannot change the values in a tuple



```python
my_tuple_list[1] = 3
```

but we can add to the list of values. This will however only work for
other tuples (try this statement with `my_list`)



```python
my_first_tuple = (1, 2, 3)
my_second_tuple = (3, 8, 7)
my_list = [12, 13, 14]
t = my_first_tuple + my_second_tuple
print(t)
```

### Sets




Sets are more like regular lists, but you can only add unique
elements. If there are duplicates, they will be ignored. Sets are
declared with curly braces:



```python
my_set = {1, 2, 3, 3}
print(my_set)
```

Set elements can be added, removed, and changed. Furthermore, sets
provide for all sorts of exciting operations which are however
beyond the scope of this course.



### Dictionaries




  Dictionaries are a datatype which
enable you to look up values based on a key, rather then index. Each
dictionary entry consists of a key-value pair. Both, key and value,
can be either a number or string.



```python
in_class_quiz = {"Domenica" : 72,
                 "Brian" : 77,
                 "George" : 95,
                 "Liz" : 81}
```

We can query the dictionary to see how individual students performed
in the quiz. However, rather than referring to the result by
a numeric index, we use the key



```python
print(in_class_quiz["Brian"])
```

Note that the key needs to be unique otherwise, you will override the
earlier definition.



```python
in_class_quiz = {"Domenica" : 72,
                 "Brian" : 77,
                 "George" : 95,
          "Brian" : 50,
                 "Liz" : 81}
print(in_class_quiz["Brian"])
```

similar to lists, we can change individual values by referring to
their key.



```python
in_class_quiz["Brian"] = 70
print(in_class_quiz["Brian"])
```
