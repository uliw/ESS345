### A note on the integer to binary conversion



This function should not use the builtin integer to binary
conversion. Rather I want you to be creative with loops and string
manipulation. You may recall that binary numbers are calculated to the base of 2.

| Base|2|2|2|2|2|
|---|---|---|---|---|---|
| Exponent|4|3|2|1|0|
|---|---|---|---|---|---|
| Expression|2^4|2^3|2^2|2^1|2^0|
|---|---|---|---|---|---|
| Result|16|8|4|2|1|
|---|---|---|---|---|---|
| Binary Number|0|1|0|0|1|

so the binary number 01001 is equal to the sum

    0 * 16 = 0
    1 * 8  = 8
    0 * 4  = 0
    0 * 2  = 0
    1 * 1  = 1
    ----------
    Total  = 9

So if you want to go the other way round, you need to first design a
loop which tests what is the biggest exponent n (i.e. 2^n) that you can
fit in the given number. In the above example, your biggest `n` would
be 3. To achieve this, you will need to use a loop in combination with
a break statement. Consider this example:



```python
a :bool= True
i :int = 0
while a:
   i  = i + 1
   if i > 10:
       break
   else:
       continue
print(i)
```

Once you know your biggest exponent (in the above example 3), you can
use a second loop which systematically counts down towards zero. This
can be done with a while loop and a counter, or with the range
function. The range function has similar syntax to the range
expressions, but rather than selecting a range from a list, it will
create a list of numbers on the fly which can be used in a for
loop. Similar to range expressions, the start value is included, but
the stop value is not.



```python
for e in range(0,5,1): # range(start, stop, step)
    print(e)
```
