## ESS345 Computational Geology: Interacting with Jupyter 1



In order to create a compute program, we need the following ingredients:

1.  A text editor which allows us to write computer code.
2.  Some sort of program which executes our code
3.  And a way to see the results of our code

There are many way to achieve the above, but typically most people use
what is commonly referred to as an "Integrated Development
Environment" or IDE. There is an enormous variety of IDE's available,
but most require a local installation. For the purpose of this course
We will use a web-based IDE, which is called `Jupyter`.  Which is
widely used in the sciences as well as in the business environment.

In this unit we you will learn:

1.  How to access a jupyter server
2.  How to upload a jupyter notebook
3.  How to execute programming statements in a jupyter notebook
4.  How to save a  jupyter notebook
5.  How to download a  jupyter notebook

Jupyter servers can run locally on you computer, are available
commercially for a fee, and some are even available for free
(e.g. googles colab). For this course, we will use a Jupyter server
which is provided by Compute Canada, and can be accessed with your
utorid.
